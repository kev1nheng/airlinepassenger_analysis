```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
%matplotlib inline
```

```python
df = pd.read_csv('airline_passengers.csv',index_col='Month',parse_dates=True)
df.index = pd.to_datetime(df.index)
```

```python
df.head(5)
```

```python
df.info()
```

```python
airline = df.dropna()
airline .plot()
```

```python
decomposition_result = seasonal_decompose(airline['Thousands of Passengers'],model = 'multiplicative')
```

```python
decomposition_result.trend.plot(title='cyclical trend',legend='Thousands of Passengers',
                               figsize=(13.5,6.5))  
```

```python
decomposition_result.seasonal.plot(title='Seasonality',legend='Thousands of Passengers', 
                                  figsize=(13.5,6.5))
```

```python
decomposition_result.resid.plot(title='Residuals',legend='Thousands of Passengers', 
                                figsize=(12,6.5))
```

```python
decomposition_result.seasonal['1949-10-01':'1950-12-01'].plot(title='Seasonal Trend per Month',
                                                             legend='Thousands of Passengers', 
                                                              figsize=(13.5,6.5))
```
